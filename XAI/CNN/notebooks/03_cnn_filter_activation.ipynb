# 03. CNN Filter Activation

각 convolution filter가 어떤 n-gram에 강하게 반응했는지, 그리고 classifier weight 기준 방향성이 무엇인지 확인한다.

```python
from pathlib import Path
import pandas as pd

cwd = Path.cwd()
if (cwd / 'xai_outputs').exists(): OUT = cwd / 'xai_outputs'
elif (cwd.parent / 'xai_outputs').exists(): OUT = cwd.parent / 'xai_outputs'
else: OUT = Path('XAI/CNN/xai_outputs')

filters = pd.read_csv(OUT / 'cnn_filter_top_ngrams.csv')
direction = pd.read_csv(OUT / 'cnn_filter_class_direction.csv')
print(OUT.resolve(), len(filters), len(direction))
```

```python
display(direction['direction_label'].value_counts().rename('count').to_frame())
display(direction.groupby(['filter_size', 'direction_label']).size().rename('count').to_frame())

top = filters.reindex(filters['target_contribution'].abs().sort_values(ascending=False).index).head(20)
display(top[['sample_id', 'target_class_name', 'filter_size', 'filter_idx', 'ngram', 'activation', 'direction_label', 'target_contribution']])
```

```python
filter_size = 3
filter_idx = int(direction[direction['filter_size'] == filter_size].iloc[0]['filter_idx'])
focus = filters[(filters['filter_size'] == filter_size) & (filters['filter_idx'] == filter_idx)]
focus = focus.sort_values('activation', ascending=False).head(15)
print('filter_size:', filter_size, 'filter_idx:', filter_idx)
display(focus[['sample_id', 'pred_label', 'target_class_name', 'ngram', 'activation', 'positive_direction', 'target_contribution']])
```
