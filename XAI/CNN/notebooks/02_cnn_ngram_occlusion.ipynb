# 02. CNN n-gram Occlusion

CNN filter size 3, 4, 5와 맞춰 n-gram 단위 occlusion 결과를 확인한다.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

cwd = Path.cwd()
if (cwd / 'xai_outputs').exists(): OUT = cwd / 'xai_outputs'
elif (cwd.parent / 'xai_outputs').exists(): OUT = cwd.parent / 'xai_outputs'
else: OUT = Path('XAI/CNN/xai_outputs')

selected = pd.read_csv(OUT / 'cnn_xai_selected_samples.csv')
ngram = pd.read_csv(OUT / 'cnn_ngram_occlusion.csv')
print(OUT.resolve(), len(ngram))
```

```python
sample_id = 'custom_3' if 'custom_3' in set(selected['sample_id']) else selected['sample_id'].iloc[0]
sample = selected[selected['sample_id'] == sample_id].iloc[0]
rows = ngram[ngram['sample_id'] == sample_id].sort_values('prob_drop', ascending=False)

print('sample_id:', sample_id)
print('text:', sample['text'])
print('tokens:', sample['tokens'])
display(rows[['n', 'start_pos', 'end_pos', 'ngram', 'base_prob', 'masked_prob', 'prob_drop']].head(20))
```

```python
display(rows.groupby('n')['prob_drop'].max().rename('max_prob_drop').to_frame())
top = rows[rows['prob_drop'] > 0].head(12).iloc[::-1].copy()
top['label'] = top['ngram'] + ' (n=' + top['n'].astype(str) + ')'
ax = top.plot.barh(x='label', y='prob_drop', legend=False, figsize=(9, 6), color='#F58518')
ax.set_title(f'n-gram occlusion: {sample_id}')
ax.set_xlabel('target probability drop')
plt.tight_layout()
```
