# 00. CNN XAI Overview

CNN XAI 산출물의 전체 구조, metadata, 대표 샘플 분포를 빠르게 확인한다. 이 노트북은 대량 재계산을 하지 않고 `xai_outputs` CSV만 읽는다.

```python
from pathlib import Path
import json
import pandas as pd

cwd = Path.cwd()
if (cwd / 'xai_outputs').exists():
    OUT = cwd / 'xai_outputs'
elif (cwd.parent / 'xai_outputs').exists():
    OUT = cwd.parent / 'xai_outputs'
else:
    OUT = Path('XAI/CNN/xai_outputs')

print('Output directory:', OUT.resolve())
print('Files:')
for path in sorted(OUT.glob('*')):
    print('-', path.name)
```

```python
metadata = json.loads((OUT / 'cnn_xai_run_metadata.json').read_text(encoding='utf-8'))
selected = pd.read_csv(OUT / 'cnn_xai_selected_samples.csv')

display(metadata)
display(selected.head())
display(selected['outcome'].value_counts(dropna=False).rename('count').to_frame())
```
