# 04. CNN Max-pooling Positions

각 filter가 문장 안에서 최종 feature로 선택한 n-gram 위치를 확인한다.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

cwd = Path.cwd()
if (cwd / 'xai_outputs').exists(): OUT = cwd / 'xai_outputs'
elif (cwd.parent / 'xai_outputs').exists(): OUT = cwd.parent / 'xai_outputs'
else: OUT = Path('XAI/CNN/xai_outputs')

selected = pd.read_csv(OUT / 'cnn_xai_selected_samples.csv')
maxpool = pd.read_csv(OUT / 'cnn_maxpool_positions.csv')
print(OUT.resolve(), len(maxpool))
```

```python
sample_id = 'custom_2' if 'custom_2' in set(selected['sample_id']) else selected['sample_id'].iloc[0]
sample = selected[selected['sample_id'] == sample_id].iloc[0]
rows = maxpool[maxpool['sample_id'] == sample_id].sort_values('target_contribution', ascending=False)

print('sample_id:', sample_id)
print('text:', sample['text'])
display(rows[['filter_size', 'filter_idx', 'max_position', 'selected_ngram', 'activation', 'target_fc_weight', 'target_contribution']].head(20))
```

```python
top = rows[rows['target_contribution'] > 0].head(12).iloc[::-1].copy()
top['label'] = top['selected_ngram'] + ' (fs=' + top['filter_size'].astype(str) + ', f=' + top['filter_idx'].astype(str) + ')'
ax = top.plot.barh(x='label', y='target_contribution', legend=False, figsize=(10, 6), color='#B279A2')
ax.set_title(f'Max-pooling contribution: {sample_id}')
ax.set_xlabel('activation x target fc weight')
plt.tight_layout()
```
